```python
df_ratings = pd.read_csv("/content/drive/MyDrive/DATATHON/dataset/title.ratings.tsv",
                         sep="\t",low_memory=False, na_values=["\\N","nan"], usecols=["tconst","averageRating"])
df_ratings.head()
```

```python
df_title_basics  = pd.read_csv("/content/drive/MyDrive/DATATHON/dataset/title.basics.tsv",
                               sep="\t",low_memory=False, na_values=["\\N","nan"], usecols=["tconst", "primaryTitle"])
df_title_basics.head()
```

```python
rating = pd.merge(df_ratings, df_title_basics, how='inner', on='tconst')
rating
```
